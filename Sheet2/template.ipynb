```python
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt
import numpy as np
```

```python
## example how to use linregress to perform a linear regression

x = np.array([0,1,2,3,4])
y = np.array([1,3,5,6,8.3])

regression = linregress(x,y)

print(regression)

print("slope:", regression.slope)
print("intercept:", regression.intercept)
```

```python
plt.plot(x,y, 'o', label='data')
plt.plot(x, regression.intercept + x*regression.slope, label='regression line')
plt.legend()
```
