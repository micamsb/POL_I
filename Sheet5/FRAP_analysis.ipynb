```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# load the data. numbers are separated by comma, hence "delimiter=','"
# the first row contains the column headers, hence skip.
data = pd.read_csv("FRAP_data.csv", delimiter=',')

# print the first lines of the data
data.head()
```

```python
# make convenient variables

# vector with all x coordinates
x = data.x
# dictionary with fluoresence levels for each time point (all but the first column)
fluorescence = {float(t):data[t] for t in data.columns[1:]}
time_points =  [float(t) for t in data.columns[1:]]
```

```python
for t in time_points:
    plt.plot(x,fluorescence[t], label=f't={t:1.2f}s')
plt.legend()
plt.xlabel('position [um]')
plt.ylabel('fluorescence intensity')
```

```python
# suitable initial condition

p = 500*np.ones_like(x)  # background level is 500
p[(x>-1.0)&(x<1.0)] = 0      # initially 0 for -1.5um<x<1.5um

plt.plot(x,p)
```
